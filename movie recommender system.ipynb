```python
#sincle title,original_title, overview, genres ,keyword, cast and crew are objects
from ast import literal_eval
def safe_literal_eval(x):
    if pd.isna(x):
        return []
    try:
        return literal_eval(x)
    except:
        return []
```

```python
df['genres'] = df['genres'].apply(safe_literal_eval)
df['keywords'] = df['keywords'].apply(safe_literal_eval)
df['cast'] = df['cast'].apply(safe_literal_eval)
df['crew'] = df['crew'].apply(safe_literal_eval)
```

```python
#print(df['genres','keywords','cast','crew']).sample
```

```python
#to extract only names from this
df['genres'] = df['genres'].apply(lambda x: [d['name'].lower().replace(" ", "_") for d in x if 'name' in d])
```

```python
df['cast'] = df['cast'].apply(lambda x: [d['name'].lower().replace(" ","_") for d in x[:3] if 'name' in d])
df['director'] = df['crew'].apply(lambda x: [d['name'].lower().replace(" ","_") for d in x if d.get('job')=='Director'][:1])
df['keywords'] = df['keywords'].apply(lambda x: [d['name'].lower().replace(" ", "_") for d in x if 'name' in d])

df[['genres','keywords','cast','crew']].sample(5)
```

```python
# Combine all tokens into a single list
df['tokens'] = df['genres'] + df['keywords'] + df['cast'] + df['director']
```

```python
df['tokens_str'] = df['tokens'].apply(lambda x: ' '.join(x))
df['soup'] = df['tokens_str'] + ' ' + df['overview'].str.lower()
# wr make soup for vecctorixation
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Create TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['soup'])
# Compute cosine similarity between movies
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_content_recommendations(title, top_n=10):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]  # exclude itself
    movie_indices = [i[0] for i in sim_scores]
    return df.iloc[movie_indices]
```

```python
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.90)

def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(v+m) * C)

df['score'] = df.apply(weighted_rating, axis=1)
```

```python
def get_hybrid_recommendations(title, top_n=10):
    # Content-based similar movies
    similar_movies = get_content_recommendations(title, top_n*2)  # get more to re-rank
    
    # Sort by weighted popularity score
    hybrid_movies = similar_movies.sort_values('score', ascending=False).head(top_n)
    
    return hybrid_movies[['title', 'vote_average', 'vote_count', 'popularity', 'score']]
```
